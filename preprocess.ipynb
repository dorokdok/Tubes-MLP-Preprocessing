```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split as splt
from scipy.stats import f_oneway
```

Melakukan Read data, dan memeriksa missing values.

```python
data = pd.read_csv("train.csv")
data2 = pd.read_csv("test.csv")
data
```

```python
data.describe(include='all')
```

Melakukan perbaikan data yang memiliki missing value. karena jumlah row yang sedikit maka kami menerapkan teknik untuk mengubah data yang missing value menjadi 0(Mengubah menjadi 0 bagia kolom yang tidak memiliki mean) dari kolomnya untuk menghindari pemangkasan data.

```python
data.replace(to_replace=['?'], value=np.nan,inplace=True)
data.replace(to_replace=np.nan, value=0,inplace=True)
data.describe(include='all')
```

Melakukan analisa untuk melakukan features selection menggunakan anova correlation dengan ketentuan pvalue > 0.05 maka feature akan didrop

```python
data.shape
```

```python
kolom = []
for col in data.columns:
    kolom.append(col)
for i in kolom[2:]:
    group = data.groupby('diagnosis')[i].apply(list)
    anova = f_oneway(*group)
    if anova[1] > 0.05:
        data.pop(i)
        data2.pop(i)
```

```python
data.shape
```

Melakukan Splitting data 70:30 untuk data train

```python
y = data.pop('diagnosis')
```

```python
x_train, x_test, y_train, y_test = splt(data,y, test_size = 0.3, random_state=1301204001)
```

Mengoutpukan menjadi file .csv Agar dapat diinputkan kedalam AutoML google

```python
train_train = pd.concat([x_train, y_train], axis=1, join='inner')
train_test = pd.concat([x_test, y_test], axis=1, join='inner')
```

```python

train_train.to_csv('processed/train-processed.csv', index=False)
train_test.to_csv('processed/train-test-processed.csv', index=False)
data2.to_csv('processed/test-processed.csv', index=False)
```
